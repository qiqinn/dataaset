```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('uts.csv')
df =df.drop(['Lat', 'Long'], axis = 1)
df
```

```python
cn= pd.DataFrame(df[df['Province/State']=='Beijing'])
id= pd.DataFrame(df[df['Country/Region']=='Indonesia'])

cn =cn.drop(['Province/State', 'Country/Region'], axis = 1)
id =id.drop(['Province/State', 'Country/Region'], axis = 1)
id
```

```python
new_df = pd.DataFrame(df.sum(numeric_only= True))
new_df.info()
```

```python
new_df.rename(columns = {0:'Global'}, inplace=True)
new_df
```

```python
new_df['China(Beijing)'] = cn.T
new_df['Indonesia'] = id.T

new_df.describe()
```

```python
new_df['Global'].plot(kind='hist', figsize=(10,7))

plt.title('Kasus Covid di dunia')


plt.show() 
```
